```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Chargement du jeu de données propre
# Assure-toi que le fichier CSV est dans le même dossier que ce script
df = pd.read_csv('passengers_clean.csv')

# 2. Configuration du style graphique général
# On privilégie un fond épuré pour faire ressortir l'information
sns.set_theme(style="whitegrid")
plt.figure(figsize=(16, 10))

# ---------------------------------------------------------
# Graphique 1 : Taux de survie par Sexe (En haut à gauche)
# ---------------------------------------------------------
plt.subplot(2, 2, 1)
# errorbar=None supprime les barres d'erreur pour une lecture plus nette
sns.barplot(x='Sex_standard', y='Survived', data=df, errorbar=None, palette=['#3498db', '#e74c3c'])
plt.title('Taux de Survie par Sexe', fontsize=14, fontweight='bold')
plt.ylabel('Probabilité de Survie')
plt.xlabel('Sexe')

# ---------------------------------------------------------
# Graphique 2 : Taux de survie par Classe Sociale (En haut à droite)
# ---------------------------------------------------------
plt.subplot(2, 2, 2)
sns.barplot(x='Pclass', y='Survived', data=df, errorbar=None, palette='viridis')
plt.title('Taux de Survie par Classe Sociale', fontsize=14, fontweight='bold')
plt.ylabel('Probabilité de Survie')
plt.xlabel('Classe du Billet (1 = 1ère classe)')

# ---------------------------------------------------------
# Graphique 3 : Distribution globale des Âges (En bas à gauche)
# ---------------------------------------------------------
plt.subplot(2, 2, 3)
sns.histplot(df['Age'], bins=30, kde=True, color='#2ecc71')
plt.title('Distribution des Âges des Passagers', fontsize=14, fontweight='bold')
plt.ylabel('Nombre de passagers')
plt.xlabel('Âge')

# ---------------------------------------------------------
# Graphique 4 : Croisement Sexe et Classe (En bas à droite)
# ---------------------------------------------------------
plt.subplot(2, 2, 4)
sns.barplot(x='Pclass', y='Survived', hue='Sex_standard', data=df, errorbar=None, palette=['#3498db', '#e74c3c'])
plt.title('Taux de Survie Croisé : Sexe et Classe', fontsize=14, fontweight='bold')
plt.ylabel('Probabilité de Survie')
plt.xlabel('Classe du Billet')
plt.legend(title='Sexe')

# 3. Ajustement de la mise en page et Sauvegarde
plt.tight_layout()
# Le script génère et sauvegarde directement une image haute résolution
plt.savefig('dashboard_titanic.png', dpi=300)

# Affichage à l'écran
plt.show()
```
